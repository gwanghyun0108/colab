```python
import random
com_list = [1,2,3,4,5,6,7,8,9]
com_num, user_num = [], []

def user_fun() :
  user = input("3개의 숫자를 입력((1~9까지) : ")
  for i in range(0,5,2) :
    user_num.append(int(user[i]))

for i in range(3) :
  r = random.choice(com_list)
  com_list.remove(r)
  com_num.append(r)

print("** 야구 게임 **")

strike, ball = 0, 0
while True :
  user_num.clear()
  user_fun()
  for i in range (3) :
    if com_num[i] == user_num[i] :
      strike = strike + 1
    elif user_num[i] in com_num :
      ball = ball + 1
  print(strike, "스트라이크", " : ", ball, "볼")
  if strike == 3 :
    print("축하합니다.")
    break
  strike, ball = 0, 0
```
